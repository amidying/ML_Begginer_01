# Multiclass Classification / Logistic Regression

The problem we are going to solve is to understand handwrite digits

```python
%matplotlib inline
import matplotlib.pyplot as plt

from sklearn.datasets import load_digits
```

To know more about the digits go to the documentation of skelarn dataset

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data[0] # printing the first one
```

```python
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
digits.target[0:5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(digits.data,digits.target,test_size=.2)
```

```python
len(X_train),len(X_test)
```

```python
type(digits)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg = LogisticRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
reg.score(X_test,y_test)
```

```python
plt.matshow(digits.images[67])
```

```python
digits.target[67]
```

```python
reg.predict([digits.data[67]])
```

### To know where my model is not working by using confusion matrix. 

```python
y_predict = reg.predict(X_test)
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test,y_predict)
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
