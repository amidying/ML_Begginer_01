### OneHotEncoder

```python
import pandas as pd
```

```python

df = pd.read_csv('homeprices.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
labelEncoderX = LabelEncoder()
```

```python
df['Encoder'] = labelEncoderX.fit_transform(df['town'])
df.head()
```

```python
df
```

```python
print(df['town'].unique())
```

```python
df['town'].value_counts()
```

```python
enc = OneHotEncoder()
```

```python
enc_data = pd.DataFrame(enc.fit_transform(df[['town']]).toarray())
```

```python
df = df.join(enc_data)
df.head()
```
