# Hypoer parameter tuning

```python
from sklearn import svm, datasets
iris =datasets.load_iris()
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df.head()
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np
```

```python
kernel = ['linear','rbf','poly','sigmoid']
C = [1,10,20]
avg_score= {}
for kval in kernel:
    for cval in C:
        cv_score = cross_val_score(svm.SVC(kernel=kval,C=cval,gamma="auto"),iris.data,iris.target,cv = 5)
        avg_score[kval+"_"+str(cval)] = np.average(cv_score)
```

```python
avg_score
```

```python
max(avg_score)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf = GridSearchCV(svm.SVC(gamma="auto"),{
    "C" : [1,10,20],
    'kernel' : ['rbf','linear','poly','sigmoid']
},cv = 5,return_train_score=False)
clf.fit(iris.data,iris.target)
clf.cv_results_
```

```python
df2 = pd.DataFrame(clf.cv_results_)
df2
```

```python
df2.mean_test_score.max()
```
